```python
import json
import os
```

```python

def build_index(json_files_directory):
    index = {}
    # Iterate over each JSON file in the directory
    for filename in os.listdir(json_files_directory):
        if filename.endswith('.json'):
            file_path = os.path.join(json_files_directory, filename)
            with open(file_path, 'r') as file:
                products = json.load(file)
                for product in products:
                    if ('id') in (product['data']['product']).keys():
                      product_id = product['data']['product']['id']
                      # Update the index with the product ID and name, mapping to the filename
                      if product_id not in index:
                          index[product_id] = filename
    with open('product_index.json', 'w') as index_file:
        json.dump(index, index_file)
    return index

build_index('all_data')
```

```python
def search_products(search_query, index_file='product_index.json'):
    # Load the index
    with open(index_file, 'r') as file:
        index = json.load(file)

    # Find the file containing the product information
    if search_query in index:
        filename = index[search_query]
        # Load the relevant JSON file and search for the product details
        with open(os.path.join('all_data', filename), 'r') as file:
            products = json.load(file)
            for product in products:
                if ('id') in (product['data']['product']).keys():
                  if str(product['data']['product']['id']) == search_query:
                      return product
    return None

# Example search
result = search_products('9853743')
print(result)
```

```python
import json
import os

def search_in_files(directory, search_key, search_value):
    matches = []
    for filename in os.listdir(directory):
        if filename.endswith('.json'):
            filepath = os.path.join(directory, filename)
            with open(filepath, 'r') as file:
                try:
                    data = json.load(file)
                    for i in data:
                      if i['data']['product'].get(search_key) == search_value:
                          matches.append(data[0]['data']['product'])
                          break
                except json.JSONDecodeError:
                    continue
    return matches

def get_info_of_products():
  directory = 'all_data'
  search_key = 'id'
  informations = []
  information_list = []
  ids_list = []
  with open('prod_data.txt', 'w+') as f_write:
    for i,r in train_data.iterrows():
      ids_list.append(str(int(r['source_product_id'])))
      ids_list.append(str(int(r['rel_product_id'])))

    ids_list = list(set(ids_list))

    for id in ids_list:
      print(id)
      new_df = []
      new_df.append(id)

      matched_products1 = search_products(id)
      categories = []
      if matched_products1:
        f_write.write('%s\t' %id)
        f_write.write('%s\n' %matched_products1)
        new_df.append(matched_products1['data']['product']['title_fa'])
        new_df.append(matched_products1['data']['product']['category']['title_fa'])
        new_df.append(matched_products1['data']['product']['images']['main']['url'][0])
        new_df.append(matched_products1['data']['product']['data_layer']['brand'])
        new_df.append(matched_products1['data']['product']['data_layer']['item_category2'])
        new_df.append(matched_products1['data']['product']['data_layer']['item_category3'])
        new_df.append(matched_products1['data']['product']['data_layer']['item_category4'])
        new_df.append(matched_products1['data']['product']['data_layer']['item_category5'])

        # new_df.append(matched_products1['data']['product']['expert_reviews'])
        if len(matched_products1['data']['product']['default_variant']) > 0:
          new_df.append(matched_products1['data']['product']['default_variant']['price']['selling_price'])
        else:
          new_df.append('Unknown')

      information_list.append(new_df)


  return information_list

information_list = get_info_of_products()
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('HooshvareLab/bert-fa-zwnj-base')
model = BertModel.from_pretrained('HooshvareLab/bert-fa-zwnj-base')

# Function to tokenize text and get BERT embeddings
def get_bert_embeddings(text):
    input_ids = tokenizer.encode(text, add_special_tokens=True, max_length=512, truncation=True)
    input_ids = torch.tensor(input_ids).unsqueeze(0)  # Batch size 1
    with torch.no_grad():
        outputs = model(input_ids)
        last_hidden_state = outputs.last_hidden_state
    # Take mean of token embeddings as sentence embeddings
    sentence_embedding = torch.mean(last_hidden_state, dim=1).squeeze().numpy()
    return sentence_embedding

# Function to normalize numerical features (price in this case)
def normalize_numerical_feature(value, min_val, max_val):
    return (value - min_val) / (max_val - min_val)


# Normalize price feature
# data['price'] = normalize_numerical_feature(data['price'], data['price'].min(), data['price'].max())

# Define weights for each text feature (title_fa and brand have higher weights)
text_features = ['title_fa', 'category', 'brand', 'cat2', 'cat3', 'cat4', 'cat5']
feature_weights = {'title_fa': 2, 'brand': 4, 'category': 2, 'cat2': 1, 'cat3': 0.6, 'cat4': 0.3, 'cat5': 0.1}

# Get BERT embeddings for text features
text_embeddings = []
for _, row in data.iterrows():
    text_embedding = np.zeros(768)  # BERT-base has 768-dimensional embeddings
    for feature in text_features:
        text_embedding += feature_weights.get(feature, 1) * get_bert_embeddings(row[feature])
    text_embeddings.append(text_embedding)

# Convert to numpy array
text_embeddings = np.array(text_embeddings)

# Concatenate price and text embeddings
# price_embeddings = np.array(data['price']).reshape(-1, 1)
# concatenated_embeddings = np.concatenate((text_embeddings, price_embeddings), axis=1)

print(text_embeddings.shape)
```

```python
train_df, test_df = train_test_split(train_data, test_size=0.2)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Concatenate, Dense
from tensorflow.keras.models import Model

def get_embeddings(product_id):
    return embeddings.get(product_id, np.zeros(768))  # Returns zero vector if ID not found

# Prepare input data
source_embeddings = np.array([get_embeddings(pid) for pid in train_df['source_product_id']])
rel_embeddings = np.array([get_embeddings(pid) for pid in train_df['rel_product_id']])
X_train = np.concatenate([source_embeddings, rel_embeddings], axis=1)
y_train = train_df['score'].values

# Build the model
input_layer = tf.keras.layers.Input(shape=(2 * 768,))
dense1 = Dense(64, activation='relu')(input_layer)
dense2 = Dense(32, activation='relu')(dense1)
output_layer = Dense(1, activation='sigmoid')(dense2)

model = Model(inputs=input_layer, outputs=output_layer)
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model
# Assuming you have a validation dataset val_df
# Prepare input data for validation set
val_source_embeddings = np.array([get_embeddings(pid) for pid in test_df['source_product_id']])
val_rel_embeddings = np.array([get_embeddings(pid) for pid in test_df['rel_product_id']])
X_val = np.concatenate([val_source_embeddings, val_rel_embeddings], axis=1)
y_val = test_df['score'].values

# Evaluate the model on validation set
mse = model.evaluate(X_val, y_val)
print("Mean Squared Error on validation set:", mse)
```
