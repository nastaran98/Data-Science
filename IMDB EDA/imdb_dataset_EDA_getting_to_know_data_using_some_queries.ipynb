#Question 1

Consider the top three genres with the highest number of movies made. How many unique movies have been made in these three genres combined? (Note that some movies may belong to multiple genres.)

```python
import ast

def safe_literal_eval(x):
    try:
        return ast.literal_eval(x)
    except (SyntaxError, ValueError):
        return None

data['genres_'] = data['genres'].apply(safe_literal_eval)

data['genres_'] = data['genres_'].apply(lambda x: [j['name'] for j in x] if x else x)

data
```

```python
movies_df = data.explode('genres_')
movies_df
```

```python
genre_counts = movies_df.groupby('genres_')['id'].size()
sorted_genres = genre_counts.sort_values(ascending=False)
sorted_genres
```

```python
top_three_genres = sorted_genres.head(3)
top_three_movies = movies_df[movies_df['genres_'].isin(top_three_genres.index)]
top_three_movies
```

```python
unique_movies_count = top_three_movies['id'].nunique()
print("Number of unique movies in the top three genres:", unique_movies_count)
```

#Question 2

Identifying film production companies and their revenues in the film industry is very important. Arrange film production companies in descending order based on the number of films they have made. What is the revenue earned from the sales of films for the fifth company in this list? (If there is a collaboration between companies, the revenue from the film is divided equally among them). Consider only the integer part of the answer and remove the decimal part.

```python
import pandas as pd

data = pd.read_csv('imdb.csv')
data.columns
```

```python
import ast

def safe_literal_eval(x):
    try:
        return ast.literal_eval(x)
    except (SyntaxError, ValueError):
        return None

data['production_companies_'] = data['production_companies'].apply(safe_literal_eval)

data['production_companies_'] = data['production_companies_'].apply(lambda x: [j['name'] for j in x] if x else x)

data
```

```python
companies_df = data.explode('production_companies_')
companies_df
```

```python
companies_count_film = companies_df.groupby('production_companies_')['id'].size()
companies_count_film_sorted = companies_count_film.sort_values(ascending=False)

companies_count_film_sorted
```

```python
# Columbia Pictures is 5th in the list

data = pd.read_csv('imdb.csv')

data['production_companies_'] = data['production_companies'].apply(safe_literal_eval)

data['production_companies_'] = data['production_companies_'].apply(lambda x: [j['name'] for j in x] if x else x)

sum_salary = 0
for i,r in data.iterrows():
  if r['production_companies_']:
    if 'Columbia Pictures' in r['production_companies_']:
      sum_salary += (r['revenue'] / len(r['production_companies_']))

print('Columbia Pictures revenue is : ', int(sum_salary))
```
