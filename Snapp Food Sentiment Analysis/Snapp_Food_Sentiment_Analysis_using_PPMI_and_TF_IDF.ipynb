```python
index_dict = {} #Dictionary to store index for each word
i = 0
for word in vocab:
    index_dict[word] = i
    i += 1

index_dict
```

```python
word_count_dict = {}
for i in vocab:
  word_count_dict[i] = 0


for i,r in data.iterrows():
  for word in list(set(r.comment_token)):
    word_count_dict[word] = word_count_dict[word] + 1

word_count_dict
```

```python
#Term Frequency
import numpy as np

def termfreq(document, word):
    N = len(document)
    occurance = len([token for token in document if token == word])
    return occurance/N

def inverse_doc_freq(documents_len,word):
    word_occurance = word_count_dict[word]
    if word_occurance == 0:
      word_occurance = 1
    # print(word, word_occurance)

    return np.log(documents_len/word_occurance)

def tf_idf(document):
    tf_idf_vec = np.zeros((len(vocab),))

    for word in document:
        tf = termfreq(document,word)
        idf = inverse_doc_freq(len(data),word)
        value = tf*idf
        tf_idf_vec[index_dict[word]] = value
    return tf_idf_vec
```

```python
vectors = []
df = data['comment_token']
vec = map(lambda x: tf_idf(x), df)

vectors = list(vec)
data['tf_idf'] = vectors

data

# print(list(i for i in (list(vec))[0] if i != 0))
```

```python
import numpy as np
import nltk
from nltk import bigrams
import itertools
import pandas as pd


def generate_co_occurrence_matrix(corpus):
    vocab = set(corpus)
    vocab = list(vocab)
    vocab_index = {word: i for i, word in enumerate(vocab)}

    # Create bigrams from all words in corpus
    bi_grams = list(bigrams(corpus))

    # Frequency distribution of bigrams ((word1, word2), num_occurrences)
    bigram_freq = nltk.FreqDist(bi_grams).most_common(len(bi_grams))

    # Initialise co-occurrence matrix
    # co_occurrence_matrix[current][previous]
    co_occurrence_matrix = np.zeros((len(vocab), len(vocab)))

    # Loop through the bigrams taking the current and previous word,
    # and the number of occurrences of the bigram.
    for bigram in bigram_freq:
        current = bigram[0][1]
        previous = bigram[0][0]
        count = bigram[1]
        pos_current = vocab_index[current]
        pos_previous = vocab_index[previous]
        co_occurrence_matrix[pos_current][pos_previous] = max(0, np.log(count * len(vocab) / count_dic[vocab[pos_current]] * count_dic[vocab[pos_previous]]))
    co_occurrence_matrix = np.matrix(co_occurrence_matrix)

    # return the matrix and the index
    return co_occurrence_matrix, vocab_index,vocab


text_data = []

for i,r in data.iterrows():
  text_data.append(r.comment_token)

# Create one list using many lists
data_ = list(itertools.chain.from_iterable(text_data))
matrix, vocab_index,vacab = generate_co_occurrence_matrix(data_)


data_matrix = pd.DataFrame(matrix, index=vocab_index,
                             columns=vocab_index)
data_matrix['vocab'] = list(vocab)
data_matrix['vocab_index'] = vocab_index
data_matrix
```
